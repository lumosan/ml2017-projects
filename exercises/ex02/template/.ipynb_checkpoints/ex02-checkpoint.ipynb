# 1 Computing the Cost Function

Fill in the `compute_loss` function below:
<a id='compute_loss'></a>


```python
def compute_mse(y, tx, w):
    '''
    Calculates the loss using mse.
    The value returned equals (2*RMSE)^2
    '''
    if len(tx.shape) == 1:
        tx = tx.reshape(-1, 1)
    if len(y.shape) == 1:
        y = y.reshape(-1, 1)
    w = np.array(w).reshape(tx.shape[1], 1)
    z = y - tx.dot(w)
    z = z.T.dot(z)
    return z[0][0] / tx.shape[0]

compute_loss = compute_mse
```

# 2 Grid Search

Fill in the function `grid_search()` below:

```python
def grid_search_naive(y, tx, w0, w1):
    '''Algorithm for grid search.'''
    losses = np.zeros((len(w0), len(w1)))
    
    for i in range(len(w0)):
        for j in range(len(w1)):
            losses[i][j] = compute_loss(y, tx, (w0[i], w1[j]))
    
    return losses
```

```python
def grid_search_naive_2(y, tx, w0, w1):
    '''Algorithm for grid search.'''
    return np.array([[compute_loss(y, tx, (w0i, w1j)) for w1j in w1] for w0i in w0])
```

# This one

```python
def gradient_descent(y, tx, initial_w, max_iters, gamma):
    """Gradient descent algorithm."""
    # Define parameters to store w and loss
    ws = [initial_w]
    losses = []
    w = initial_w
    
    for n_iter in range(max_iters):
        # Compute gradient and loss
        gradient = compute_gradient(y, tx, w)
        loss = compute_loss(y, tx, w)
        
        # Update w by gradient
        aux = w - gamma * gradient
        w = aux
        
        # store w and loss
        ws.append(w)
        losses.append(loss)
        print("Gradient Descent({bi}/{ti}): loss={l}, w0={w0}, w1={w1}".format(
              bi=n_iter, ti=max_iters - 1, l=loss, w0=w[0], w1=w[1]))

    return losses, ws
```

```python
def compute_stoch_gradient(y, tx, w):
    """Compute a stochastic gradient from just few examples n and their corresponding y_n labels."""
    # ***************************************************
    # INSERT YOUR CODE HERE
    # TODO: implement stochastic gradient computation.It's same as the gradient descent.
    # ***************************************************
    n = np.random.randint(0, np.shape(y)[0])
    aux1 = y[n] - tx[n].dot(w)
    return -(tx[n].dot(aux1))


def stochastic_gradient_descent(
        y, tx, initial_w, batch_size, max_iters, gamma):
    """Stochastic gradient descent algorithm."""
    # ***************************************************
    # INSERT YOUR CODE HERE
    # TODO: implement stochastic gradient descent.
    # ***************************************************

    ws = [initial_w]
    losses = []
    w = initial_w
    
    for n_iter in range(max_iters):
        # ***************************************************
        # TODO: compute gradient and loss
        # ***************************************************
        gradient = compute_stoch_gradient(y, tx, w)
        loss = compute_loss(y, tx, w)
        
        # ***************************************************
        # TODO: update w by gradient
        # ***************************************************
        aux = w - gamma * gradient
        w = aux
        
        # store w and loss
        ws.append(w)
        losses.append(loss)
        print("Gradient Descent({bi}/{ti}): loss={l}, w0={w0}, w1={w1}".format(
              bi=n_iter, ti=max_iters - 1, l=loss, w0=w[0], w1=w[1]))

    return losses, ws
```

Modify the function `compute_loss(y, tx, w)` for the Mean Absolute Error cost function [here](#compute_loss)

```python
def compute_loss(y, tx, w):
    """Calculate the loss.

    You can calculate the loss using mse or mae.
    """
    return .5*np.sum(np.abs(y-tx.dot(w)))/np.shape(y)[0]
```
